```python
import pyautogui
import time
import datetime
```

```python
# pyautogui.FAILSAFE = False

print("Started at {}".format(datetime.datetime.now().time()))

while True:
    day_of_week = datetime.date.today().weekday() # 0 is Monday, 6 is Sunday
    time_of_day = datetime.datetime.now().time()
    
    time.sleep(5)

    ### Move left and right, press 1 twice to attack with first attack
    for laps in range(0,3):
        for i in range(0,120):

            pyautogui.keyDown('right')
            time.sleep(400/1000)
            pyautogui.keyUp('right')

            pyautogui.keyDown('left')
            time.sleep(400/1000)
            pyautogui.keyUp('left')

            pyautogui.press('1')
            time.sleep(100/1000)
            pyautogui.press('1')

            print(i)
        time.sleep(4)
        pyautogui.press('4')
        time.sleep(3)
        pyautogui.press('6')
        time.sleep(10/1000)
        pyautogui.press('1')
        time.sleep(10/1000)
        pyautogui.press('1')
        time.sleep(10/1000)
        
    
        
    print("Ended at {}".format(datetime.datetime.now().time()))
    break 
```
